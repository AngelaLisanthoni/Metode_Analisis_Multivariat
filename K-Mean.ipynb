### Perhitungan K-Mean

```python
import numpy as np

def distance(p, q):
    return np.sqrt(np.sum((p - q) ** 2))

def k_means(data, k, initial_centroids):
    centroids = initial_centroids.copy()
    
    iteration = 1
    
    # Cetak jarak Euclidean pertama dengan titik pusat awal
    print("\nTabel Jarak Awal:")
    print("   x   |   y   | " + " | ".join([f"C{i+1}" for i in range(k)]))
    print("-" * 29)
    for i, point in enumerate(data):
        distances = [distance(point, centroid) for centroid in initial_centroids]
        print(f" {point[0]:^5} | {point[1]:^5} | " + " | ".join([f"{distances[j]:^4}" for j in range(k)]))
    print()
    
    while True:
        print(f"Iterasi {iteration}")
        iteration += 1
        
        # Inisialisasi cluster
        clusters = [[] for _ in range(k)]
        
        # Hitung jarak dan atributkan data ke centroid terdekat
        for point in data:
            distances = [distance(point, centroid) for centroid in centroids]
            closest_centroid_idx = np.argmin(distances)
            clusters[closest_centroid_idx].append(point)
        
        # Hitung centroid baru
        new_centroids = []
        for cluster in clusters:
            new_centroid = np.mean(cluster, axis=0)
            new_centroids.append(new_centroid)
        
        # Cetak jarak per iterasi
        #for i, centroid in enumerate(new_centroids):
           # distances = [distance(point, centroid) for point in data]
           # print(f"Jarak centroid ke-{i+1} (mean): {np.mean(distances)}")
            #print(f"Jarak centroid ke-{i+1} (min): {np.min(distances)}")
        
        # Cetak tabel jarak
        print("\nTabel Jarak:")
        print("   x   |   y   | " + " | ".join([f"C{i+1}" for i in range(k)]))
        print("-" * 29)
        for i, point in enumerate(data):
            distances = [distance(point, centroid) for centroid in new_centroids]
            print(f" {point[0]:^5} | {point[1]:^5} | " + " | ".join([f"{distances[j]:^4}" for j in range(k)]))
        print()
        
        # Konvergensi
        if np.allclose(centroids, new_centroids):
            break
        
        # Update centroid
        centroids = new_centroids
        
        # Cetak cluster saat ini
        #for i, cluster in enumerate(clusters):
           # print(f"Cluster {i+1}: {cluster}")
        
        print(f"Centroid baru: {centroids}\n")
    
    return clusters

# Contoh penggunaan dengan data baru
data = np.array([
    [2, 0],
    [1, 2],
    [2, 2],
    [3, 2],
    [2, 3],
    [3, 3],
    [2, 4],
    [3, 4],
    [4, 4],
    [3, 5]
])

k = 2

# Meminta input titik awal dari pengguna
initial_centroids = []
for i in range(k):
    x = float(input(f"Masukkan koordinat x untuk centroid ke-{i+1}: "))
    y = float(input(f"Masukkan koordinat y untuk centroid ke-{i+1}: "))
    initial_centroids.append([x, y])

initial_centroids = np.array(initial_centroids)

clusters = k_means(data, k, initial_centroids)
```

```python
def distance(p, q):
    return np.sqrt(np.sum((p - q) ** 2))

def k_means(data, k, initial_centroids):
    num_variables = data.shape[1]  # Get the number of variables in the data
    centroids = initial_centroids.copy()
    iteration = 1

    # Print the initial Euclidean distances with the initial centroids
    print("\nInitial Distance Table:")
    header = " | ".join([f"C{i+1}" for i in range(k)])
    print(" | ".join([f"x{i+1}" for i in range(num_variables)]) + " | " + header)
    print("-" * (6 * num_variables + 3 + 6 * k))
    for i, point in enumerate(data):
        distances = [distance(point, centroid) for centroid in initial_centroids]
        print(" | ".join([f"{coord:^6}" for coord in point]) + " | " + " | ".join([f"{distances[j]:^6.2f}" for j in range(k)]))
    print()

    while True:
        print(f"Iteration {iteration}")
        iteration += 1

        # Initialize clusters
        clusters = [[] for _ in range(k)]

        # Calculate distances and assign data to the nearest centroid
        for point in data:
            distances = [distance(point, centroid) for centroid in centroids]
            closest_centroid_idx = np.argmin(distances)
            clusters[closest_centroid_idx].append(point)

        # Calculate new centroids
        new_centroids = []
        for cluster in clusters:
            new_centroid = np.mean(cluster, axis=0)
            new_centroids.append(new_centroid)


        # Print distance table
        print("\nDistance Table:")
        print(" | ".join([f"x{i+1}" for i in range(num_variables)]) + " | " + header)
        print("-" * (6 * num_variables + 3 + 6 * k))
        for i, point in enumerate(data):
            distances = [distance(point, centroid) for centroid in new_centroids]
            print(" | ".join([f"{coord:^6}" for coord in point]) + " | " + " | ".join([f"{distances[j]:^6.4f}" for j in range(k)]))
        print()

        # Convergence check
        if np.allclose(centroids, new_centroids):
            break

        # Update centroids
        centroids = new_centroids


        print(f"New Centroids: {centroids}\n")

    return clusters

data = np.array([
    [1, 125000, 1000000],
    [3, 150000, 1250000],
    [5, 200000, 2500000],
    [6, 220000, 2750000],
    [1, 100000, 1000000]
])

k = 2

initial_centroids = []
for i in range(k):
    centroid = []
    for j in range(data.shape[1]):
        value = float(input(f"Enter coordinate {j+1} for centroid {i+1}: "))
        centroid.append(value)
    initial_centroids.append(centroid)

initial_centroids = np.array(initial_centroids)

clusters = k_means(data, k, initial_centroids)
```

```python
import numpy as np
from sklearn.cluster import KMeans

# Input data
data = np.array([[1, 125000, 1000000],
                 [3, 150000, 1250000],
                 [5, 200000, 2500000],
                 [6, 220000, 2750000],
                 [1, 100000, 1000000]])

# Initialize K-means with the desired number of clusters
kmeans = KMeans(n_clusters=2)

# Fit the data to the K-means model
kmeans.fit(data)

# Get the cluster labels for each data point
labels = kmeans.labels_

# Get the cluster centers
centers = kmeans.cluster_centers_

# Print the cluster labels and centers
for i in range(len(data)):
    print("Data point {}: label={}, center={}".format(i+1, labels[i], centers[labels[i]]))
```
